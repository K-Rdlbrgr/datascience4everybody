## QUESTIONS

### 1. Select al the countries of America

````python
# This gives us a string of Booleans where it is indicated which countries include AMER in their region name
american = countries_df['Region'].str.contains('AMER')

# Now we can use this as a filter for the dataframe
countries_df[american]
````

### Select all the Eruopean Countries

````python
# Again we need to define a filter
european = countries_df['Region'].isin(['WESTERN EUROPE', 'EASTERN EUROPE'])

# We use european as a filter
countries_df[european]

# Unfortunately this does not work because after the 'EASTERN EUROPE' string there are a bunch of spaces -> 'EASTERN EUROPE            '
````

````python
# We need to clean the data and specifically the region column: Map goes through the list and applies the str.strip function which gets rid off the trailing white space
countries_df['Region'].map(str.strip)

# Now we need to assign this change to the original dataframe
countries_df['Region'] = countries_df['Region'].map(str.strip)

# We use the functionality again
european = countries_df['Region'].isin(['WESTERN EUROPE', 'EASTERN EUROPE'])
countries_df[european]
````

## Filtering columns

````python
# Filter soecific columns out of the dataframes
countries_df.filter(items=['Country', 'Population'])

# This does the same as
countries_df[['Country', 'Population']]
````

## Re-indexing

````python
# The set_index method re-indexes the dataframe with the specified column
countries_df.set_index('Country')
````

````python
# But if we not store it, the dataframe will not save the change when you call it again
countries_df.head()
````

````python
# We need to save it in order to keep it that way
# Before we do that, there is some more white space to clean. There is a space after each country name
countries_df['Country'] = countries_df['Country'].map(str.strip)
countries_df = countries_df.set_index('Country')

# This enables us to call rows by their index which is now a specific country
countries_df.loc['Portugal']
````

## Filtering Rows

### Getting all countries that start with ```P```

````python
# Regex is a regular expression -> Pattern recognition (^ represents starting of a string; $ represents the end of a string)
# The following line willl give all countries starting with P
countries_df.filter(regex='^P', axis='rows')
````

## Sorting

### Based on Rows

````python
# This sorts our data based on the index in descending fashion (some from the back)
countries_df.sort_index(ascending=False)
````

### Based on Columns

````python
# Sorting the countries by their population, starting with the biggest
countries_df.sort_values('Population', ascending=False)
````

## Chaining functionalities

````python
# Most likely we want to combain several filters and seletions in order to get some valuable output
# In order to do so we can simply chain them together with '.' and to write it in a clear fashion we use '\' to seperate them in lines
countries_df.sort_values('Population', ascending=False) \
    .head(10) \
    .filter(items=['Region', 'Population', 'Area (sq. mi.)'])
````

## Grouping

Grouping is used to split the dataframe in different buckets which we can perform individual actions on and then display each result afterwards

### Calculate the Population of each region

````python
# We need the groupby function for this to group the countries by their regions and then save it
regions = countries_df.groupby('Region')

# Now we can sum it
regions.sum()

````

````python
# Since this is hard to read, let's modify the format
top_regions = regions[['Population', 'Area (sq. mi.)']].sum() \
    .sort_values('Population', ascending=False) / 1_000_000

# The last division with the underscores moves the dot forewards to make it more readable, but be aware of it since the unit is not displayed in the header or anywhere else
top_regions
````

````python
# So let's actually rename the columns to solve this issue
top_regions.columns = ['Population (mm)', 'Area (sq. mi.) (mm)']

# Print it out
top_regions
````

### Which region is the most dense?

````python
# We calculate each value by taking the mean() instead of the sum() from each region
regions[['Pop. Density (per sq. mi.)']].mean() \
    .sort_values('Pop. Density (per sq. mi.)', ascending=False) \
    # We round the values
    .round(1)

````

## Advanced Aggregation on Grouping

### Getting the population and GDP per capita per region

This is more complex since we need the sum of the population and the mean of the GDP per capita

````python
# We call the base for the calculation which is regions and call the agg(regation) function
# For this we are using numpy functions with np.sum and np.mean
# Then we round the values
# And we sort the regions by the GDP per capita
regions.agg({ 'Population': np.sum, 'GDP ($ per capita)': np.mean }) \
    .round(1) \
    .sort_values('GDP ($ per capita)', ascending=False)
````

## Boolean Indexing

### Getting the population and GDP per capita per region

We often need toa sk questions based on conditions.

````python
# We use a condition and store the boolean column in a variable
pop_greater_than_100 = countries_df['Population'] > 100_000_000
pop_greater_than_100
````

````python
# Once defined, the condition can be used as an index to filter based on the condition
countries_df[pop_greater_than_100]
````

````python
# We are foltering countries that have a GDP below 1000
gdp_lower_than_1000 = countries_df['GDP ($ per capita)'] <= 1000

# Let's combine the two conditions with a '&' and filter the data
countries_df.loc[gdp_lower_than_1000 & pop_greater_than_100]
````

## Plotting

Plot the GDP per capita of the top ten countries

````python
# Let's assign the column first
column = 'GDP ($ per capita)'
# We need the double brackets to create a Dataframe instead of a Series, sort the values and store the top twenty in a new variable
top_ten_countries_df = countries_df[[column]] \
    .sort_values(column, ascending=False) \
    .head(20)
top_ten_countries_df
````

````python
# Let's plot this sub dataframe
top_ten_countries_df.plot(kind='bar')
````
